```python
import datetime

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## упражнение 1 

```python
bars_length = np.random.randint(-10, 10, 10)
bars_length.sort()

yticks_labels = ['some label: {}'.format(i) for i in range(10)]
labels = ['some text: {}'.format(i) for i in range(10)]
figsize = (15, 10)

def plot_horizontal_bar_1(bars_length, labels, yticks_labels, figsize):
    fig = plt.figure(figsize = (15, 10))
    ax = fig.gca()
    for i in range(10):
        ax.barh(i, bars_length[i], label = labels[i], color = 'lime', alpha = 0.7)
        if bars_length[i] < 0:
            ax.annotate(labels[i], xy = (0,i), xytext = (0.4, i), fontsize = 25)
        else:
            ax.annotate(labels[i], xy = (0,i), xytext = (bars_length[i] + 0.1, i), fontsize = 25)
            
    ax.set_yticks(np.arange(0, 10))
    ax.set_yticklabels(yticks_labels, fontsize = 18, fontstyle = 'italic')
    
plot_horizontal_bar_1(bars_length, labels, yticks_labels, figsize)
```

## упражнение 2

```python
dti = pd.date_range('2013-01-01', '2020-12-31', freq='2W')
probabilities_in_time = np.random.random((10, len(dti)))
probabilities_in_time = probabilities_in_time / probabilities_in_time.sum(axis=0)
probabilities_in_time = pd.DataFrame(probabilities_in_time).T
probabilities_in_time.index = dti
figsize = (20, 5)
colors = ['', 'green', 'yellowgreen','yellow', 'orange','darkorange','orangered', 'tomato',  'red','brown']


def plot_time_probabilities(pit, fs):
    plt.figure(figsize=fs)
    plt.plot(pit[0], label=f'Probability: {0}', color='#003391')
    plt.fill_between(pit.index, pit[0], color='#003391') 
    for i in range(1, 10):
        pit[i] += pit[i-1]
        plt.plot(pit[i], label=f'Probability: {i}', color=colors[i])
        plt.fill_between(pit.index, pit[i-1], pit[i], color=colors[i])
    plt.legend(prop={'size': 16}, loc='lower left')
    plt.grid(True)
    plt.show()
    
plot_time_probabilities(probabilities_in_time, figsize)
```
